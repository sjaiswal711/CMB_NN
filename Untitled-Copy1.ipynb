```python
# To save map.dat occurance, centre_pix, neighbors_weights
import numpy as np
temp_format = "%.8f"
centre_pix_format = "%d"
occurance_format = "%d"

fmt = [occurance_format] + [centre_pix_format] + [temp_format]+[temp_format] # Split format_string by tabs

nside = 256
npix = 12*nside**2

array_map = np.zeros((npix, 4))

array_map[:, 1] = np.arange(npix)

print(array_map)

np.savetxt("2_map.dat",array_map,fmt = fmt)

print("Array map saved to array_map.dat")

```

```python
import numpy as np
index = [1,2,4,5,6]
cen_pix = []
Temp = []
for idx in index:
    print(idx)
    filename = f"{idx}month.dat"
    data = np.loadtxt(filename)
    cen_pix = data[:,1]
    Temp = data[:,2]
    for i in range(len(cen_pix)):
        pixel = cen_pix[i]
        index = int(pixel)
        map_data[index,2] = map_data[index,2]+ Temp[i] # add the weights later divide it by occurance
        map_data[index,0] = map_data[index,0] + 1  # occurance ++
```
